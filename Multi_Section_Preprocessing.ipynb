```python
import pandas as pd

# Selecting relevant columns (Make sure all datasets exist)
income_data = income_data[['Household_Code', 'monthly_income', 'annual_income', 'sec_monthly_income', 'sec_annual_income', 'other_sec_income']]
housing_data = housing_data[['Household_Code', 'Occupancy_Status', 'Number_of_Rooms', 'Ownership_of_Internet', 'Ownership_of_Mobile_Phone', 'Ownership_of_Vehicle']]
wash_data = wash_data[['Household_Code', 'Water_Treatment', 'Monthly_Water_Cost', 'Water_Cost_Reasonable']]
solid_waste_data = solid_waste_data[['Household_Code', 'Payment_for_Waste_Services', 'Frequency_of_Bin_Clearance']]
assets_data = assets_data[['Household_Code', 'Ownership_Agricultural_Land', 'Ownership_Residential_Building', 'Irrigation_Status_of_Land']]
durable_data = durable_data[['Household_Code', 'Number_of_Items_Owned']]
education_data = education_data[['Household_Code', 'Expense_Fees', 'Expense_Transport', 'Expense_Supplies']]

# Convert 'Household_Code' to a common data type (to avoid misalignment)
```

```python
def optimize_dtypes(df):
    for col in df.columns:
        col_type = df[col].dtype
        if col_type == 'object':
            df[col] = df[col].astype('category')
        elif col_type == 'float64':
            df[col] = df[col].astype('float32')
        elif col_type == 'int64':
            df[col] = df[col].astype('int32')
    return df

def process_df(df, key):
    df = df.drop_duplicates(subset=key).set_index(key)
    return optimize_dtypes(df)

# Process all DataFrames
income_data = process_df(income_data, 'Household_Code')
housing_data = process_df(housing_data, 'Household_Code')
wash_data = process_df(wash_data, 'Household_Code')
solid_waste_data = process_df(solid_waste_data, 'Household_Code')
assets_data = process_df(assets_data, 'Household_Code')
durable_data = process_df(durable_data, 'Household_Code')
education_data = process_df(education_data, 'Household_Code')
```

```python
# Start with the first DataFrame
merged_df = income_data

# List of remaining DataFrames to merge
dfs_to_merge = [
    housing_data,
    wash_data,
    solid_waste_data,
    assets_data,
    durable_data,
    education_data
]

# Merge iteratively
for df in dfs_to_merge:
    merged_df = merged_df.join(df, how='left')
    del df  # Free memory immediately

# Reset index if needed (optional)
merged_df = merged_df.reset_index()
```

```python
# Replace all missing values with 0 based on your instruction
merged_df.fillna(0, inplace=True)

# Special cases for categorical columns (convert to proper types)
merged_df['Water_Treatment'] = merged_df['Water_Treatment'].astype(int)
merged_df['Ownership_Agricultural_Land'] = merged_df['Ownership_Agricultural_Land'].astype(int)
merged_df['Ownership_Residential_Building'] = merged_df['Ownership_Residential_Building'].astype(int)
merged_df['Irrigation_Status_of_Land'] = merged_df['Irrigation_Status_of_Land'].astype(int)
```

```python
# Convert to appropriate numeric types
int8_cols = ['Occupancy_Status', 'Ownership_of_Internet', 'Ownership_of_Mobile_Phone',
             'Ownership_of_Vehicle', 'Water_Cost_Reasonable']

float32_cols = ['monthly_income', 'annual_income', 'sec_monthly_income',
                'sec_annual_income', 'other_sec_income', 'Monthly_Water_Cost',
                'Payment_for_Waste_Services', 'Frequency_of_Bin_Clearance',
                'Expense_Fees', 'Expense_Transport', 'Expense_Supplies']

merged_df[int8_cols] = merged_df[int8_cols].astype('int8')
merged_df[float32_cols] = merged_df[float32_cols].astype('float32')
```

```python
# Create composite features
merged_df['total_income'] = merged_df['monthly_income'] + merged_df['sec_monthly_income']
merged_df['total_expenses'] = (merged_df['Monthly_Water_Cost'] +
                              merged_df['Expense_Fees'] +
                              merged_df['Expense_Transport'] +
                              merged_df['Expense_Supplies'])

merged_df['asset_score'] = (merged_df['Ownership_of_Internet'] +
                           merged_df['Ownership_of_Mobile_Phone'] +
                           merged_df['Ownership_of_Vehicle'] +
                           merged_df['Number_of_Items_Owned'])

merged_df['housing_quality'] = (merged_df['Number_of_Rooms'] * 2) + \
                              merged_df['Ownership_Residential_Building']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define features list
features = [
    'total_income',
    'total_expenses',
    'asset_score',
    'housing_quality',
    'Ownership_Agricultural_Land',
    'Irrigation_Status_of_Land',
    'Water_Treatment',
    'Payment_for_Waste_Services'
]

# Prepare X and y from your DataFrames
X = merged_df[features]
y = depv['poverty_status']  # 0 for non-poor, 1 for poor

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets (80% train, 20% test) with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Define a dictionary of models (SVM removed)
model=RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

# Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc
import numpy as np

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix with a vibrant colormap
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='magma', xticklabels=['Non-Poor', 'Poor'], yticklabels=['Non-Poor', 'Poor'])
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('Actual', fontsize=12)
plt.title('Confusion Matrix - Poverty Prediction', fontsize=14, fontweight='bold')
plt.show()

# Feature importance plot (for Random Forest) with different colors for each bar
feature_importance = model.feature_importances_
sorted_idx = np.argsort(feature_importance)
colors = plt.cm.viridis(np.linspace(0, 1, len(features)))  # Generate unique colors

plt.figure(figsize=(8, 6))
plt.barh(np.array(features)[sorted_idx], feature_importance[sorted_idx], color=colors)
plt.xlabel('Feature Importance', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.title('Feature Importance - Random Forest', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

# ROC Curve with improved colors
y_prob = model.predict_proba(X_test)[:, 1]  # Probability estimates for class 1 (Poor)
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='black', linestyle='--', lw=1.5)  # Diagonal reference line
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curve - Poverty Prediction', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
